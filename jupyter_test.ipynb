## 5. Test Package Management

Let's check what packages are available and test installing new ones if needed.

````python
# Check some key installed packages
import pkg_resources

key_packages = ['numpy', 'pandas', 'matplotlib', 'jupyter', 'opencv-python', 'scikit-learn']

print("Key Package Versions:")
for package in key_packages:
    try:
        version = pkg_resources.get_distribution(package).version
        print(f"  {package}: {version}")
    except pkg_resources.DistributionNotFound:
        print(f"  {package}: Not installed")

# Test importing some data science libraries
try:
    import cv2
    print(f"\nOpenCV version: {cv2.__version__}")
except ImportError:
    print("\nOpenCV not available")

try:
    import sklearn
    print(f"Scikit-learn version: {sklearn.__version__}")
except ImportError:
    print("Scikit-learn not available")
````

## 6. Verify Kernel Information

Let's display information about the Python kernel and available computing resources.

````python
# Display kernel and system information
import os
import multiprocessing
import psutil

print("=== Kernel Information ===")
print(f"Python executable: {sys.executable}")
print(f"Python version: {sys.version}")
print(f"Current working directory: {os.getcwd()}")

print("\n=== System Resources ===")
print(f"CPU cores: {multiprocessing.cpu_count()}")
print(f"Available memory: {psutil.virtual_memory().total / (1024**3):.2f} GB")
print(f"Available disk space: {psutil.disk_usage('.').free / (1024**3):.2f} GB")

print("\n=== Environment Variables ===")
conda_env = os.environ.get('CONDA_DEFAULT_ENV', 'Not set')
print(f"Conda environment: {conda_env}")
print(f"PATH: {os.environ.get('PATH', 'Not set')[:100]}...")

# Test if we're running in Jupyter
try:
    from IPython import get_ipython
    ipython = get_ipython()
    print(f"\nRunning in IPython: {ipython is not None}")
    if ipython:
        print(f"IPython version: {IPython.__version__}")
except:
    print("\nNot running in IPython environment")
````

#### **Option 2: Create Project-Specific Environments (Best Practice)**

- **Cons:** Requires creating new environments for each project

### 🔧 **How to Create New Environments for Specific Projects:**

```bash
# Create a new environment for a machine learning project
conda create -n ml_project python=3.13 pandas numpy matplotlib scikit-learn jupyter

# Create environment for web development
conda create -n web_dev python=3.13 flask django requests jupyter

# Activate specific environment
conda activate ml_project
```

### 💡 **My Recommendation:**
For learning and small projects, use your **base environment**. For larger projects or when working with different Python versions, create dedicated environments.
